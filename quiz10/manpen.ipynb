```python
import networkx as nx
from itertools import product



grid=""".....
.S-7.
.|.|.
.L-J.
.....
""".splitlines()

grid="""..F7.
.FJ|.
SJ.L7
|F--J
LJ...""".splitlines()

with open("manpen.txt") as f:
    grid = f.readlines()

grid = """FF7FSF7F7F7F7F7F---7
L|LJ||||||||||||F--J
FL-7LJLJ||||||LJL-77
F--JF--7||LJLJIF7FJ-
L---JF-JLJIIIIFJLJJ7
|F|F-JF---7IIIL7L|7|
|FFJF7L7F-JF7IIL---7
7-L-JL7||F7|L7F-7F7|
L.L7LFJ|||||FJL7||LJ
L7JLJL-JLJLJL--JLJ.L""".replace("I", ".").splitlines()

grid = list(map(str.strip, grid))
cols = len(grid[0])
rows = len(grid)

assert all(len(line) == cols for line in grid)

def to_idx(x, y):
    #return x + y * rows
    return (x,y)

def from_idx(idx):
    #return idx % rows, idx // rows
    return idx

G = nx.Graph()
for x, y in product(range(cols), range(rows)):
    G.add_node(to_idx(x, y))

def add_edge(G, x,y, dx, dy):
    if 0 <= x + dx < cols and 0 <= y + dy < rows:
        G.add_edge(to_idx(x, y), to_idx(x + dx, y + dy))

def add_edges(G, x,y, d1, d2):
    add_edge(G, x, y, *d1)
    add_edge(G, x, y, *d2)

start = None

def neighbors(grid, x, y):
    W = (-1, 0)
    E = (1, 0)
    N = (0, -1)
    S = (0, 1)

    cell = grid[y][x]
    n = []

    if cell == "-":
        n = [W, E]
    elif cell == "|":
        n = [N, S]
    elif cell == "L":
        n = [N, E]
    elif cell == "J":
        n = [N, W]
    elif cell == "7":
        n = [S, W]
    elif cell == "F":
        n = [S, E]
    elif cell == "S":
        n = [N, S, W, E]
    elif cell == ".":
        pass
    else:
        assert False, cell

    return [(x + dx, y + dy) for (dx, dy) in n if 0 <= x + dx < cols and 0 <= y + dy < rows]

for x, y in product(range(cols), range(rows)):
    for neigh in neighbors(grid, x, y):
        if (x, y) in neighbors(grid, *neigh):
            G.add_edge(to_idx(x, y), to_idx(*neigh))

    if grid[y][x] == "S":
        start = to_idx(x, y)
   

assert start is not None
#layers=list(nx.bfs_layers(G, start))

#for i, layer in enumerate(layers):
#    #if i > 1 and any(G.has_edge(x, start) for x in layer):
#    print(any(G.has_edge(x, start) for x in layer))


cycle = nx.find_cycle(G, start)
print(len(cycle) / 2)
```

```python
nodes = set()
for u, v in cycle:
    nodes.add(u)
    nodes.add(v)

mgrid = [[" "] * (cols + 2) for _ in range(rows + 2)]
for x, y in product(range(cols), range(rows)):
    if grid[y][x] == ".":
        mgrid[y+1][x+1] = "."

for (x, y) in nodes:
    mgrid[y+1][x+1] = "*" if grid[y][x] != "-" else "-"


stack = [(0,0)]
while len(stack) > 0:
    x, y = stack.pop()
    mgrid[y][x] = "_"

    for dx, dy in [(x, y-1), (x, y+1), (x-1, y), (x+1, y)]:
        if  dx < 0 or dx > cols + 1 or  dy < 0 or dy > rows + 1:
            continue

        if mgrid[dy][dx] in ["*", "_", "-"]:
            continue

        mgrid[dy][dx] = "_"
        stack.append((dx, dy))

count = 0
for y, line in enumerate(mgrid):
    inside = 0
    for x, c in enumerate(line):
        if c == "*":
            inside += 2 if grid[y-1][x-1] == "|" else 1
            continue

        if c == "-":
            continue

        assert inside % 2 == 0, line

        if c == "." and inside % 4 == 2:
            count += 1
            mgrid[y][x] = "I"


print("\n".join("".join(l) for l in mgrid))
print(sum(sum(x == "." for x in line) for line in mgrid))
print(count)
```
