```python
import numpy as np

with open("manpen.txt") as f:
    text = f.read()

example = """O....#....
O.OO#....#
.....##...
OO.#O....O
.O.....O#.
O.#..O.#.#
..O..#O..O
.......O..
#....###..
#OO..#....
"""

def move_north(matrix):
    idx = list(zip(*np.nonzero( matrix == "O" )))
    for r, c in idx:
        nr = r
        while nr > 0 and matrix[nr-1, c] == ".":
            nr -= 1

        matrix[r, c] = "."
        matrix[nr, c] = "O"

    return matrix

def beam_load(matrix):
    sum = 0
    for r in range(matrix.shape[0]):
        s = matrix.shape[0] - r 
        sum += s * np.sum(matrix[r, :] == "O")

    return sum

def quiz2(text):
    matrix = np.array([list(line) for line in text.splitlines()])
    iter = dict()

    N = 1000000000
    i = 0
    while i < N:
        for _ in range(4):
            matrix = move_north(matrix).T [:, ::-1]

        start = iter.setdefault(matrix.tobytes(), i)
        if start < i:
            period = i - start
            i += ((N - i - 1) // period) * period 

        i += 1

    return beam_load(matrix)


def quiz(text):
    matrix = np.array([list(line) for line in text.splitlines()])
    matrix = move_north(matrix)
    return beam_load(matrix)

assert quiz(example) == 136
assert quiz(text) == 106517

assert quiz2(example) == 64
assert quiz2(text) == 79723
```
